```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# Scatter Plot
X_data = np.random.rand(100)*100
```

```python
Y_data = np.random.rand(100)*100
```

```python
plt.scatter(X_data, Y_data, c="blue", s=150, alpha=0.2)
```

```python
# Line Plots
years = [2006 + x for x in range(16)]
```

```python
weights = np.random.randint(45, 100, 16)
```

```python
plt.plot(years, weights, c="r", alpha=0.3, lw=3, linestyle="-")
```

```python
# Bar Chart
programming_languages = ["C", "python", "C++", "java", "ruby"]
```

```python
scores = np.random.randint(45, 120, 5)
```

```python
plt.bar(programming_languages, scores, color="red", align="edge", width=0.5, edgecolor="crimson", lw=2)
```

```python
# Histogram
ages = np.random.normal(10,1.5,1000)
```

```python
plt.hist(ages, bins=20)
```

```python
# Pie Chart
langs = ["Python", "C++", "C", "Ruby", "Java"]
votes = np.random.randint(200, 500, 5)
explodes = [0, 0, 0, 0, 0.2]
plt.pie(votes, labels=langs, explode=explodes, startangle=90, pctdistance=1.3, autopct="%.2f%%")
plt.title('Programming Language Votes')
```

```python
# Box & Wiskher Plot
heights = np.random.normal(172, 8, 300)
plt.boxplot(heights)
```

```python
num1 = np.linspace(0, 10, 25)
num2 = np.linspace(10, 200, 25)
num3 = np.linspace(200, 210, 25)
num4 = np.linspace(210, 230, 25)

data = np.concatenate((num1, num2, num3, num4))
plt.boxplot(data)
```

```python
years = [2014+x for x in range(8)]
income = [55, 56, 62, 61, 72, 72, 73, 75]
income_ticks = list(range(50, 81, 2))
```

```python
# Basics of Plot Customization
plt.title("Income of John (in USD)", fontsize=15)
plt.xlabel("Year")
plt.ylabel("Yearly Income in USD")
plt.yticks(income_ticks, [f"${x}k" for x in income_ticks])
plt.plot(years, income)
```

```python
# Legends with Multiple Line Plots
tesla = np.random.randint(100, 120, 7)
apple = np.random.randint(90, 150, 7)
alphabet = np.random.randint(120, 150, 7)
plt.plot(tesla, label="tesla")
plt.plot(apple, label="apple")
plt.plot(alphabet, label="alphabet")
plt.legend(loc="lower center")
```

```python
# Legends with Pie Chart 
votes = [10, 2, 5, 16, 22]
people = ["A", "B", "C", "D", "E"]
plt.pie(votes, labels=None)
plt.legend(labels=people)
```

```python
# Multiple Figures 
x1, y1 = np.random.random(100), np.random.random(100)
x2, y2 = np.arange(100), np.arange(100)
```

```python
plt.figure(1)
plt.scatter(x1, y1)
plt.figure(2)
plt.plot(x2, y2)
```

```python
# Subplots
x = np.arange(1, 101)  

fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("Sine Wave")

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("Cosine Wave")

axs[1, 0].plot(x, np.random.random(100))
axs[1, 0].set_title("Random Function")

axs[1, 1].plot(x, np.log1p(x))
axs[1, 1].set_title("Log Function")
```

```python
# 3D Plotting

ax = plt.axes(projection="3d")

x = np.random.random(100)
z = np.random.random(100)
y = np.random.random(100)

ax.scatter(x, y, z)
```
